### 1)
sequence-to-sequence: predict time series e.g. music,weather

sequence-to-vector: predict movie prefs based on history

vector-to-sequence: image captions, generate data based on params
### 2)
better to translate after hearing the whole sentence, instead of word for word (out of context)
### 3)
cnn to classify features and then rnn for time axis classification
### 4)
dynamic sentence lenghts when translating
### 5)
padding or dynamic_rnn
### 6)
each layer on a different gpu

```python
import sklearn
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import os
```

```python
reber_grammar = [
    [[1,'B']],
    [[2,'T'],[3,'P']],
    [[2,'S'],[4,'X']],
    [[3,'T'],[5,'V']],
    [[6,'S'],[3,'X']],
    [[4,'P'],[6,'V']],
    [[-1,'E']]
]

embedded_reber_grammar = [
    [[1,'B']],
    [[2,'T'],[3,'P']],
    [[4,'T']],
    [[4,'P']],
    [[-1,'E']]
]

def gen_text(reber):
    state = 0
    text = ""
    while state != -1:
        idx = np.random.randint(len(reber[state]))
        text = text + reber[state][idx][1]
        state = reber[state][idx][0]
    return text

#gen (hopefully) false text by shuffeling input
def gen_false_text(text):
    text = list(text)
    np.random.shuffle(text)
    return "".join(text)

#gen 50/50
def gen_data(n_samples, ratio):
    X = [[]]*n_samples
    for i in range(n_samples - np.int(np.floor(n_samples/2))):
        txt = gen_text(embedded_reber_grammar)
        embed = txt[:2] + gen_text(reber_grammar) + txt[2:]
        X[i] = [embed,1]
        
        chars="BEPSTVX"
        index = np.random.randint(len(embed))
        good_char = embed[index]
        bad_char = np.random.choice(sorted(set(chars) - set(good_char)))
        bad = embed[:index] + bad_char + embed[index + 1:]
        if np.random.random() <= ratio:
        #only cange 1 char (harder)
            X[np.int(np.floor(n_samples/2)) + i] = [bad,0]
        else:
        #shuffle        
            X[np.int(np.floor(n_samples/2)) + i] = [gen_false_text(embed),0]
    return X

#str to one-hot
def str_one_hot(string, n_chars, cat = "BTPXVSE "):
    X = np.zeros((n_chars, len(cat)))
    i=0
    for char in string:
        X[i, cat.find(char)] = 1
        i+=1
    return X

#gen one-hot dataset from X
def to_one_hot(X):
    max_len = [len(string) for string in X[:,0]]
    max_len = np.max(max_len)
    X_new = np.zeros((X.shape[0],) + str_one_hot(X[0,0], max_len).shape)
    for i in range(X.shape[0]):
        X_new[i] = str_one_hot(X[i,0], max_len)
    return X_new

#str to num
def to_num_vec(X):
    chars = "BTPXVSE "
    char_to_index = {char: index for index, char in enumerate(chars)}
    X_new = np.copy(X[:,0])
    X_new = [[char_to_index[char] for char in string] for string in X_new]
    return X_new

def build_model(vocab_size, embedding_dim, rnn_units, batch_size):
    model = tf.keras.Sequential([
        tf.keras.layers.Embedding(vocab_size, embedding_dim),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(rnn_units, recurrent_initializer='glorot_uniform', return_sequences=True)),
        tf.keras.layers.Bidirectional(tf.keras.layers.GRU(rnn_units,recurrent_initializer='glorot_uniform')),
        tf.keras.layers.Dense(vocab_size, activation='relu'),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    return model
```

```python
n_samples = 40000
X = np.array(gen_data(n_samples, 0.7))
np.random.shuffle(X)
y = X[:,1].astype('int32')
#X_hot = to_one_hot(X)
X_num_vecs = to_num_vec(X)

train_data = keras.preprocessing.sequence.pad_sequences(X_num_vecs,
                                                        value=7,#" "
                                                        padding='post',
                                                        maxlen=31)

model = build_model(
  vocab_size = 8,
  embedding_dim=256,
  rnn_units=31,
  batch_size=64)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

x_val = train_data[30000:37000]
partial_x_train = train_data[:30000]
x_test = train_data[37000:40000]

y_val = y[30000:37000]
partial_y_train = y[:30000]
y_test = y[37000:40000]

train_dataset = tf.data.Dataset.from_tensor_slices((partial_x_train, partial_y_train)).shuffle(20000).batch(1024)
test_dataset = tf.data.Dataset.from_tensor_slices((x_test, y_test)).shuffle(20000).batch(1024)
val_dataset = tf.data.Dataset.from_tensor_slices((x_val, y_val)).shuffle(20000).batch(1024)

checkpoint_path = "training_2/cp-{epoch:04d}.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)

cp_callback = tf.keras.callbacks.ModelCheckpoint(checkpoint_path,
                                                 save_weights_only=True,
                                                 verbose=1)


history = model.fit(train_dataset,
                    epochs=42,
                    validation_data=val_dataset,
                    verbose=1,
                    callbacks = [cp_callback])
```

```python
n_samples = 40000
X = np.array(gen_data(n_samples, 1))
np.random.shuffle(X)
y = X[:,1].astype('int32')
#X_hot = to_one_hot(X)
X_num_vecs = to_num_vec(X)

train_data = keras.preprocessing.sequence.pad_sequences(X_num_vecs,
                                                        value=7,#" "
                                                        padding='post',
                                                        maxlen=31)
x_val = train_data[30000:37000]
partial_x_train = train_data[:30000]
x_test = train_data[37000:40000]
y_val = y[30000:37000]
partial_y_train = y[:30000]
y_test = y[37000:40000]

train_dataset = tf.data.Dataset.from_tensor_slices((partial_x_train, partial_y_train)).shuffle(20000).batch(1024)
test_dataset = tf.data.Dataset.from_tensor_slices((x_test, y_test)).shuffle(20000).batch(1024)
val_dataset = tf.data.Dataset.from_tensor_slices((x_val, y_val)).shuffle(20000).batch(1024)

#train further on harder errors only
latest = tf.train.latest_checkpoint(checkpoint_dir)
model.load_weights(latest)
checkpoint_path = "training_3/cp-{epoch:04d}.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)

cp_callback = tf.keras.callbacks.ModelCheckpoint(checkpoint_path,
                                                 save_weights_only=True,
                                                 verbose=1)
history2 = model.fit(train_dataset,
                        epochs=42,
                        validation_data=val_dataset,
                        verbose=1,
                        callbacks = [cp_callback])
```

```python
results = model.evaluate(x_test, y_test)
print(results)
history_dict = history.history
```

```python
acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']
loss = history_dict['loss']
val_loss = history_dict['val_loss']
loss2 = history2.history['loss']
val_loss2 = history2.history['val_loss']

epochs = range(1, len(acc) + 1)

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.plot(epochs, loss2, 'ro', label='Training loss 2')
plt.plot(epochs, val_loss2, 'r', label='Validation loss 2')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
acc2 = history2.history['accuracy']
val_acc2 = history2.history['val_accuracy']
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.plot(epochs, acc2, 'ro', label='Training acc 2')
plt.plot(epochs, val_acc2, 'r', label='Validation acc 2')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```
